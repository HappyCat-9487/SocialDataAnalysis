```python
import pandas as pd 
data = pd.read_csv('Data_1.csv')
data.head()
```

```python
print(data['Category'].value_counts())
```

```python
print(len(data['Category']))
```

```python
import matplotlib.pyplot as plt
data['Category'].value_counts().hist(bins=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

data['Category'].value_counts().hist(bins=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

x = np.arange(len(data.groupby(by = 'Category')))
plt.bar(x, data['Category'].value_counts(), color= 'blue')
plt.xticks(x, data.groupby(by = 'Category'))
plt.xlabel('Category')
plt.ylabel('Occurence')
plt.title('Crime Occurences')
plt.show()
```
