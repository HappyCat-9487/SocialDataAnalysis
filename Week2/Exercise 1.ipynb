```python
import numpy as np
import urllib.request
with urllib.request.urlopen('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data1.tsv') as f:
    data1 = f.read().decode('utf-8')
print(data1)

with urllib.request.urlopen('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data2.tsv') as f:
    data2 = f.read().decode('utf-8')
print(data2)

with urllib.request.urlopen('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data3.tsv') as f:
    data3 = f.read().decode('utf-8')
print(data3)

with urllib.request.urlopen('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/data4.tsv') as f:
    data4 = f.read().decode('utf-8')
print(data4)

data1 = data1.split( )
data2 = data2.split( )
data3 = data3.split( )
data4 = data4.split( )

data1_x = []
for i in range(0,len(data1),2):
    data1_x.append(int(data1[i]))
print(data1_x)

data2_x = []
for i in range(0,len(data2),2):
    data2_x.append(int(data2[i]))
print(data2_x)

data3_x = []
for i in range(0,len(data3),2):
    data3_x.append(int(data3[i]))
print(data3_x)

data4_x = []
for i in range(0,len(data4),2):
    data4_x.append(int(data4[i]))
print(data4_x)



data1_y = []
for i in range(1,len(data1),2):
    data1_y.append(float(data1[i]))
print(data1_y)

data2_y = []
for i in range(1,len(data2),2):
    data2_y.append(float(data2[i]))
print(data2_y)

data3_y = []
for i in range(1,len(data3),2):
    data3_y.append(float(data3[i]))
print(data3_y)

data4_y = []
for i in range(1,len(data4),2):
    data4_y.append(float(data4[i]))
print(data4_y)

data1_x_mean = np.mean(data1_x)
data1_y_mean = np.mean(data1_y)

data2_x_mean = np.mean(data2_x)
data2_y_mean = np.mean(data2_y)

data3_x_mean = np.mean(data3_x)
data3_y_mean = np.mean(data3_y)

data4_x_mean = np.mean(data4_x)
data4_y_mean = np.mean(data4_y)

print(data1_x_mean)
print(data1_y_mean)
print(data2_x_mean)
print(data2_y_mean)
print(data3_x_mean)
print(data3_y_mean)
print(data4_x_mean)
print(data4_y_mean)
```

```python
data_x_all = np.array(data1_x+data2_x+data3_x+data4_x)
print(data_x_all)
var_x = np.var(data_x_all)
print("{:0.3f}".format(var_x))

data_y_all = np.array(data1_y+data2_y+data3_y+data4_y)
print(data_y_all)
var_y = np.var(data_y_all)
print("{:0.3f}".format(var_y))

data_pc = np.corrcoef(data_x_all,data_y_all)
np.set_printoptions(precision=3)
print(data_pc)

from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(data1_x,data1_y)
print("{:0.2f}".format(slope))
print("{:0.2f}".format(intercept))
slope, intercept, r_value, p_value, std_err = stats.linregress(data2_x,data2_y)
print("{:0.2f}".format(slope))
print("{:0.2f}".format(intercept))
slope, intercept, r_value, p_value, std_err = stats.linregress(data3_x,data3_y)
print("{:0.2f}".format(slope))
print("{:0.2f}".format(intercept))
slope, intercept, r_value, p_value, std_err = stats.linregress(data4_x,data4_y)
print("{:0.2f}".format(slope))
print("{:0.2f}".format(intercept))
```

```python
import matplotlib.pyplot as plt
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(data1_x,data1_y)
a1, b1 = slope, intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(data2_x,data2_y)
a2, b2 = slope, intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(data3_x,data3_y)
a3, b3 = slope, intercept
slope, intercept, r_value, p_value, std_err = stats.linregress(data4_x,data4_y)
a4, b4 = slope, intercept

x1 = np.linspace(0, 20)
y1 = np.array(a1*x1+b1)
y2 = np.array(a2*x1+b2)
y3 = np.array(a3*x1+b3)
y4 = np.array(a4*x1+b4)

f, axs = plt.subplots(2, 2, sharex='all', sharey='all')
axs[0, 0].scatter(data1_x, data1_y)
axs[0, 0].plot(x1, y1)
axs[0, 1].scatter(data2_x, data2_y)
axs[0, 1].plot(x1, y2)
axs[1, 0].scatter(data3_x, data3_y)
axs[1, 0].plot(x1, y3)
axs[1, 1].scatter(data4_x, data4_y)
axs[1, 1].plot(x1, y4)
axs[0, 0].set_title('Data 1')
axs[0, 1].set_title('Data 2')
axs[1, 0].set_title('Data 3')
axs[1, 1].set_title('Data 4')
plt.show()
```
