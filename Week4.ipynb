```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import folium

map_hooray = folium.Map(location=[37.77919, -122.41914],
                    zoom_start = 13) # Uses lat then lon. The bigger the zoom number, the closer in you get
map_hooray # Calls the map to display
```

```python
data = pd.read_csv('Data_1.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data[(data['Date'] >= pd.to_datetime('06/01/2017')) & (data['Date'] <= pd.to_datetime('07/31/2017'))]
data.set_index('Category')
data.index = data['Category']
focus_data = data.loc['DRUG/NARCOTIC', :]
print(focus_data)
```

```python
#print(focus_data.iloc[0, 12])
#print(focus_data.iloc[0, 11])

map_hooray = folium.Map(location=[37.77919, -122.41914], zoom_start = 13)

for i in range(len(focus_data)):
    folium.Marker([focus_data.iloc[i][12], focus_data.iloc[i][11]], popup='San Francisco').add_to(map_hooray)

map_hooray
```

```python
data_2 = pd.read_csv('Data_1.csv')
```

```python
data_2.index = data_2['Category']
focus_data_2 = data_2.loc['SEX OFFENSES, NON FORCIBLE', :]
print(focus_data_2)
```

```python
from folium import plugins
from folium.plugins import HeatMap


map_hooray = folium.Map(location=[37.77919, -122.41914],
                    zoom_start = 13) 


# List comprehension to make out list of lists
heat_data = [[row['Y'],row['X']] for index, row in focus_data_2.iterrows()]

# Plot it on the map
HeatMap(heat_data, radius=30, blur=10).add_to(map_hooray)

# Display the map
map_hooray
```
