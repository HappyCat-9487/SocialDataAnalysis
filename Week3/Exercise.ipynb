```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

data_male = pd.read_csv('Male.txt')
data_female = pd.read_csv('Female.txt')

#print(data_male.columns)
data = data_male.append(data_female, ignore_index = True)
print(data)
print(stats.ttest_ind(data_male['steps'], data_female['steps']))

f, axs = plt.subplots(1, 2, sharex='all', sharey='all')
axs[0].hist(data_male['steps'], bins='fd')
axs[1].hist(data_female['steps'], bins='fd', color='red')
plt.show()
```

```python
my_rho = np.corrcoef(data_female['steps'], data_female['bmi'])
print(my_rho)
```

```python
my_rho_male = np.corrcoef(data_male['steps'], data_male['bmi'])
print(my_rho_male)
```

```python
x1 = np.arange(0, len(data_male))
x2 = np.arange(0, len(data_female))
f, axs = plt.subplots(1, 2, sharex='all', sharey='all', figsize=(15,10))
axs[0].scatter(data_male['steps'], data_male['bmi'], color='#f6756d', marker='x')
axs[1].scatter(data_female['steps'], data_female['bmi'], color='#10bdc3', marker='^')
plt.show()

plt.figure(figsize=(10,10))
plt.scatter(data['steps'], data['bmi'], marker='P')

plt.show()
```
